```python
import time
structure = [5, 8, 8, 8, 4]
structure[0] = len(test_x[0])
structure[-1] = len(test_y[0])
print(structure)
weights = []

activation_selector = 2

for i in range(len(structure)-1):
    w = 2*np.random.rand(structure[i], structure[i+1])-1
    weights.append(np.array(w))
#print(weights)

x = np.array([1, 0, 1])
y = [0, 1, 0, 1, 0, 1, 0]

lr = 0.1 + 0.9*(activation_selector==0)
#lr = 1.0
delta = np.zeros(len(structure)-1)
dw = np.zeros(len(structure)-1)
a = np.zeros(len(structure))
relu_val = 0.1
#print(np.matmul(a, weights[0]))
#takes input returns y_hat 

def differ(fx, val = 0):
    if val == 0:
        return fx*(1-fx)
    elif val == 1:
        return (1 - (fx ** 2))
    elif val == 2:
        return (1.0)*(fx>0)+(relu_val)*(fx<0)
    
def sigmoid(fx, val = 0):
    if val == 0:
        return 1.0 / (1.0 + exp(-fx))
    elif val == 1:
        return np.tanh(fx)
    elif val == 2:
        return fx*(fx>0) + relu_val*fx*(fx<0)

def forward_prop(inpt, delta, a, dw):
    #weights = best_weights.copy()
    #weights = np.load('best_weights.npy') 
    out = inpt
    a.append(inpt)
    for i in range(len(weights)):
        out = np.dot(out, weights[i])
        out = sigmoid(out,val = activation_selector)
        delta.append(out)
        a.append(np.array(out))
        dw.append(out)
    y_hat = out
    #print(len(a))
    return a[-1], delta, a, dw

#forward_prop(x)
#takes delta and backpropagates 

def backward_prop(dlta, delta, a, dw):
    m= len(dlta)
    delta[-1] = dlta.copy()
    dw[-1] = (1/m)*np.dot(a[-2].T, delta[-1])
    #print(dw[-1].shape)
    cur_delta = dlta.copy()
    for i in range(len(weights)-2,-1,-1):
        delta[i] = np.multiply(delta[i+1].dot(weights[i+1].T), differ(a[i+1],val = activation_selector))
        dw[i] = (1/m)*np.dot(a[i].T, delta[i])#je layer er delta tar ager layer er activation er shathe gun for eact example
    for i in range(len(weights)):
        #print(weights[i].shape, dw[i].shape)
        weights[i] = weights[i] - lr*dw[i]
        
def train(lr):
    #lr = 0.1
    prev = 1
    start_time = time.time()
    new_time = start_time
    for epoch in range(500):
        delta = []
        a = []
        dw = []
        y_hat,delta, a, dw = forward_prop(train_x, delta, a, dw)
        #print(y_hat)
        error = 0.5*(train_y-y_hat)**2
        #print(error)
        if(epoch%100 == 0):
            lr -= 0.001
            new_time = time.time()-new_time
            print("current error =", round(error.sum(), 12), " time = ", round(new_time, 6), "lr = ", round(lr, 6))
            if(error.sum()<=prev):
                prev = error.sum()
                np.save('best_weights.npy', weights)
            new_time = time.time()
        #print(y_hat.shape)
        dlta = (y_hat-train_y)*differ(a[-1], val = activation_selector)
        backward_prop(dlta, delta, a, dw)
    elapsed_time = time.time() - start_time
    print("total time needed for training = ", str(elapsed_time))
    

train(lr)

def test():
    miss = 0
    predicted, _, _, _ = forward_prop(train_x, [], [], [])
    #print(predicted.shape)
    #print(predicted[:10])
    for i in range(len(predicted)):
        val = -1
        pred = -1
        for j in range(len(predicted[0])):
            if(predicted[i][j]>val):
                val = predicted[i][j]
                pred = j
        if(train_y[i][pred] != 1.00):
            miss+=1
            #print("test y = ", test_y[i], "predicted = ",predicted[i][pred])
    print("training data total", len(train_x), "missed = ", miss)
    
    
    miss = 0
    predicted, _, _, _ = forward_prop(test_x, [], [], [])
    #print(predicted.shape)
    #print(predicted[:10])
    for i in range(len(predicted)):
        val = -1
        pred = -1
        for j in range(len(predicted[0])):
            if(predicted[i][j]>val):
                val = predicted[i][j]
                pred = j
        if(test_y[i][pred] != 1.00):
            miss+=1
            #print("test y = ", test_y[i], "predicted = ",predicted[i][pred])
    print("testing data total", len(test_x), "missed = ", miss)
    
test()
```

```python
import pandas as pd
import numpy as np


df = pd.read_csv('W1data.csv')
#df.head() 
print("len df = ", len(df))
# Get labels
y = df[['Cultivar 1', 'Cultivar 2', 'Cultivar 3']].values

# Get inputs; we define our x and y here.
X = df.drop(['Cultivar 1', 'Cultivar 2', 'Cultivar 3'], axis = 1)
X.shape, y.shape # Print shapes just to check
X = X.values

import matplotlib
import matplotlib.pyplot as plt
# SciKitLearn is a useful machine learning utilities library
import sklearn
# The sklearn dataset module helps generating datasets
import sklearn.datasets
import sklearn.linear_model
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

# Now we define all our functions

def softmax(z):
    #Calculate exponent term first
    exp_scores = np.exp(z)
    return exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

def softmax_loss(y,y_hat):
    # Clipping value
    minval = 0.000000000001
    # Number of samples
    m = y.shape[0]
    # Loss formula, note that np.sum sums up the entire matrix and therefore does the job of two sums from the formula
    loss = -1/m * np.sum(y * np.log(y_hat.clip(min=minval)))
    return loss

def loss_derivative(y,y_hat):
    return (y_hat-y)

def tanh_derivative(x):
    return (1 - np.power(x, 2))

# This is the forward propagation function
def forward_prop(model,a0):
    
    #Start Forward Propagation
    
    # Load parameters from model
    W1, b1, W2, b2, W3, b3 = model['W1'], model['b1'], model['W2'], model['b2'], model['W3'],model['b3']
    
    # Do the first Linear step 
    # Z1 is the input layer x times the dot product of the weights + our bias b
    z1 = a0.dot(W1) + b1
    
    # Put it through the first activation function
    a1 = np.tanh(z1)
    
    # Second linear step
    z2 = a1.dot(W2) + b2
    
    # Second activation function
    a2 = np.tanh(z2)
    
    #Third linear step
    z3 = a2.dot(W3) + b3
    
    #For the Third linear activation function we use the softmax function, either the sigmoid of softmax should be used for the last layer
    a3 = softmax(z3)
    
    #Store all results in these values
    cache = {'a0':a0,'z1':z1,'a1':a1,'z2':z2,'a2':a2,'a3':a3,'z3':z3}
    return cache

# This is the BACKWARD PROPAGATION function
def backward_prop(model,cache,y):

    # Load parameters from model
    W1, b1, W2, b2, W3, b3 = model['W1'], model['b1'], model['W2'], model['b2'],model['W3'],model['b3']
    
    # Load forward propagation results
    a0,a1, a2,a3 = cache['a0'],cache['a1'],cache['a2'],cache['a3']
    
    # Get number of samples
    m = y.shape[0]
    
    # Calculate loss derivative with respect to output
    dz3 = loss_derivative(y=y,y_hat=a3)

    # Calculate loss derivative with respect to second layer weights
    dW3 = 1/m*(a2.T).dot(dz3) #dW2 = 1/m*(a1.T).dot(dz2) 
    #print("W3 shape", W3.shape,"del W3 shape", dW3.shape)
    # Calculate loss derivative with respect to second layer bias
    db3 = 1/m*np.sum(dz3, axis=0)
    
    # Calculate loss derivative with respect to first layer
    dz2 = np.multiply(dz3.dot(W3.T) ,tanh_derivative(a2))
    
    print( dz3.shape, "dot with ", (W3.T).shape ," got = " , (dz3.dot(W3.T)).shape,"then multiply with",  a2.shape)
    
    
    # Calculate loss derivative with respect to first layer weights
    dW2 = 1/m*np.dot(a1.T, dz2)
    
    # Calculate loss derivative with respect to first layer bias
    db2 = 1/m*np.sum(dz2, axis=0)
    
    dz1 = np.multiply(dz2.dot(W2.T),tanh_derivative(a1))
    
    dW1 = 1/m*np.dot(a0.T,dz1)
    
    db1 = 1/m*np.sum(dz1,axis=0)
    
    # Store gradients
    grads = {'dW3':dW3, 'db3':db3, 'dW2':dW2,'db2':db2,'dW1':dW1,'db1':db1}
    return grads

#TRAINING PHASE
def initialize_parameters(nn_input_dim,nn_hdim,nn_output_dim):
    # First layer weights
    W1 = 2 *np.random.randn(nn_input_dim, nn_hdim) - 1
    #print("W1.shape = " , W1.shape)
    # First layer bias
    b1 = np.zeros((1, nn_hdim))
    
    # Second layer weights
    W2 = 2 * np.random.randn(nn_hdim, nn_hdim) - 1
    #print("W2.shape = " , W2.shape)
    
    # Second layer bias
    b2 = np.zeros((1, nn_hdim))
    W3 = 2 * np.random.rand(nn_hdim, nn_output_dim) - 1
    print("W3.shape = " , W3.shape)
    b3 = np.zeros((1,nn_output_dim))
    
    
    # Package and return model
    model = { 'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2,'W3':W3,'b3':b3}
    return model

def update_parameters(model,grads,learning_rate):
    # Load parameters
    W1, b1, W2, b2,b3,W3 = model['W1'], model['b1'], model['W2'], model['b2'],model['b3'],model["W3"]
    
    # Update parameters
    W1 -= learning_rate * grads['dW1']
    b1 -= learning_rate * grads['db1']
    W2 -= learning_rate * grads['dW2']
    b2 -= learning_rate * grads['db2']
    W3 -= learning_rate * grads['dW3']
    b3 -= learning_rate * grads['db3']
    
    # Store and return parameters
    model = { 'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2, 'W3':W3,'b3':b3}
    return model
def predict(model, x):
    # Do forward pass
    c = forward_prop(model,x)
    #get y_hat
    y_hat = np.argmax(c['a3'], axis=1)
    return y_hat
def calc_accuracy(model,x,y):
    # Get total number of examples
    m = y.shape[0]
    # Do a prediction with the model
    pred = predict(model,x)
    # Ensure prediction and truth vector y have the same shape
    pred = pred.reshape(y.shape)
    # Calculate the number of wrong examples
    error = np.sum(np.abs(pred-y))
    # Calculate accuracy
    return (m - error)/m * 100
losses = []
def train(model,X_,y_,learning_rate, epochs=20, print_loss=False):
    # Gradient descent. Loop over epochs
    for i in range(0, epochs):

        # Forward propagation
        cache = forward_prop(model,X_)
        #a1, probs = cache['a1'],cache['a2']
        # Backpropagation
        
        grads = backward_prop(model,cache,y_)
        # Gradient descent parameter update
        # Assign new parameters to the model
        model = update_parameters(model=model,grads=grads,learning_rate=learning_rate)
    
        # Pring loss & accuracy every 100 iterations
        if print_loss and i % 10 == 0:
            a3 = cache['a3']
            print('Loss after iteration',i,':',softmax_loss(y_,a3))
            y_hat = predict(model,X_)
            y_true = y_.argmax(axis=1)
            print('Accuracy after iteration',i,':',accuracy_score(y_pred=y_hat,y_true=y_true)*100,'%')
            losses.append(accuracy_score(y_pred=y_hat,y_true=y_true)*100)
    return model
np.random.seed(0)
# This is what we return at the end
model = initialize_parameters(nn_input_dim=13, nn_hdim= 5, nn_output_dim= 3)
model = train(model,X,y,learning_rate=0.07,epochs=4500,print_loss=True)
plt.plot(losses)
```
